```python
reviews_arr = []
dates_arr = []
url = 'https://thesmartlocal.com/read/insurance-policy-singapore/'
r = requests.get(url)

soup = BeautifulSoup(r.content, 'html.parser')

qa_dict = {}

for section in soup.findAll('h3'):
    nextNode = section
    if section.text[0].isdigit():
        qa_dict[section.text] = ''
        while True:
            if nextNode != None:
                nextNode = nextNode.find_next_sibling()
                if nextNode and nextNode.name == 'hr':
                    ans = ''
                    while True:
                        if nextNode != None:
                            nextNode = nextNode.find_next_sibling()
                            if nextNode and nextNode.name == 'p':
                                ans += nextNode.text
                            elif nextNode and nextNode.name == 'h4':
                                while True:
                                    if nextNode != None:
                                        nextNode = nextNode.find_next_sibling()
                                        if nextNode and nextNode.name == 'p':
                                            ans += nextNode.text
                                        else:
                                            break

                            else:
                                break
                    qa_dict[section.text] = ans
                    break
        else:
            break
# print(qa_dict)
```

```python
df = pd.DataFrame(columns = ['question', 'answer'])
for i, x in qa_dict.items():
    df = df.append({'question': i, 'answer': x}, ignore_index=True)
df
unclean_csv_data = df.to_csv('UNCLEANDATA.csv', index = True) 
```
