```python
import pandas as pd
import numpy as np
```

## Import the dataset that you want to clean

```python
# labels to be last column
df = pd.read_csv("train.csv") 
df
```

## Check no. of records

```python
df.info()
```

## Check for NA in all columns

```python
#check for na
df.isnull().sum(axis = 0)
```

## Remove NA

```python
# drop those col with > 70% missing values
pct_null = df.isnull().sum() / len(df)
missing_features = pct_null[pct_null > 0.70].index
df.drop(missing_features, axis=1, inplace=True)
```

```python
missing_features   #columns that are dropped
```

```python
#input missing values with mean for the other columns
df = df.fillna(df.mean())
```

## Check for features that are unique to each customer

```python
df.nunique()
```

## Remove columns where features are unique to each customer

```python
# get columns where nunique count == no. of records in dataframe
columns = []
for col in df.columns:
    if len(df[col]) == len(df[col].unique()):
        columns.append(col)

columns
```

```python
# drop columns where nunique num == no. of records --> meaning columns where it is unique to each customer for eg. Id
df = df.drop(columns = columns)
```

Check that columns are removed

```python
df.head()
```

## Label Encoding

Change textual data to number eg. Yes/No to 1/0 or Age 10-20, Age 21-30, Age 31-40 to 0,1,2 respectively

```python
# add all column names into a list called 'feature'
feature=[]
for col in df.columns:
    feature.append(col)
```

```python
#check value type for each column
for i in feature:
    print(df[i].dtypes)
```

```python
# if valye type is object (which means it is a text), label encoder will encode it into numerical values
from sklearn.preprocessing import LabelEncoder
lb_style = LabelEncoder()

for i in feature:
    if df[i].dtypes=='object':
        df[i]= lb_style.fit_transform(df[i])
```

Check that textual values are encoded into numerical values

```python
df.head()
```

## Change last column to 'labels'

```python
df.columns = [*df.columns[:-1], 'labels']
df
```

## Convert labels to binary classification 

`Only if your labels are multiclass`

```python
#changing multi-class labels to binary
labels =[]
for i in df['labels']:
    if i==8:
        labels.append(0) # No risk
    else:
        labels.append(1) # Risk
```

```python
#drop existing label column with new label
df.drop(columns=['labels'],inplace=True)
df['labels']=labels
```

## Export cleaned training dataset

```python
df.to_csv('train_cleaned.csv', index=False)
```
