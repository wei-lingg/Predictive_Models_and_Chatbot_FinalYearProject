```python
import pandas as pd
import numpy as np
```

## Import the dataset that you want to clean

```python
# labels to be last column
df = pd.read_csv("train.csv") 
```

## Check no. of records

```python
df.info()
```

## Check for NA in all columns

```python
#check for na
df.isnull().sum(axis = 0)
```

## Remove NA

```python
df = df.dropna()
```

## Check for features that are unique to each customer

```python
df.nunique()
```

## Remove columns where features are unique to each customer

```python
# get columns where nunique count == no. of records in dataframe
columns = []
for col in df.columns:
    if len(df[col]) == len(df[col].unique()):
        columns.append(col)

columns
```

```python
# drop columns where nunique num == no. of records --> meaning columns where it is unique to each customer for eg. customerID
df = df.drop(columns = columns)
```

Check that columns are removed

```python
df.head()
```

## Data standardisation/scaling

```python
# before standardisation
df.head()
```

```python
#scale continuous variables only (i.e. feature 0  to feature 6)
from sklearn import preprocessing

#change columns_to_scale to column headers where values in that column is continuous
columns_to_scale = ['feature_0', 'feature_1','feature_2','feature_3','feature_4','feature_5','feature_6'] 
standardized_X = preprocessing.scale(df[columns_to_scale])
standardized_features = pd.DataFrame(standardized_X)
df[columns_to_scale] = standardized_features

# after standardisation
df.head()
```

## Label Encoding

Change textual data to number eg. Yes/No to 1/0 or Age 10-20, Age 21-30, Age 31-40 to 0,1,2 respectively

```python
# add all column names into a list called 'feature'
feature=[]
for col in df.columns:
    feature.append(col)
```

```python
#check value type for each column
for i in feature:
    print(df[i].dtypes)
```

```python
# if valye type is object (which means it is a text), label encoder will encode it into numerical values
from sklearn.preprocessing import LabelEncoder
lb_style = LabelEncoder()

for i in feature:
    if df[i].dtypes=='object':
        df[i]= lb_style.fit_transform(df[i])
```

Check that textual values are encoded into numerical values

```python
df.head()
```

## Export cleaned training dataset

```python
df.to_csv('train_cleaned.csv', index=False)
```
