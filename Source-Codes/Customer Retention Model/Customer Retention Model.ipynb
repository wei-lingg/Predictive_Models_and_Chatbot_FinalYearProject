## Data Standardization/Scaling
`Standardization refers to shifting the distribution of each attribute to have a mean of zero and a standard deviation of one`

```python
#for continuous variables only (i.e. feature 0  to feature 6)
from sklearn import preprocessing

features_to_scale = train[['feature_0', 'feature_1','feature_2','feature_3','feature_4','feature_5','feature_6']]
standardized_X = preprocessing.scale(features_to_scale)
standardized_features = pd.DataFrame(standardized_X)

train['feature_0'] = standardized_features[0]
train['feature_1'] = standardized_features[1]
train['feature_2'] = standardized_features[2]
train['feature_3'] = standardized_features[3]
train['feature_4'] = standardized_features[4]
train['feature_5'] = standardized_features[5]
train['feature_5'] = standardized_features[6]

train.head()
```

## Split into Train, Validation and Test

```python
from sklearn.model_selection import train_test_split

X = train.drop('labels', 1)
y = train['labels']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)
```

## Balancing the imbalanced data

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
X_train, y_train = oversample.fit_resample(X_train, y_train)
```

```python
start = dt.datetime.now()
clf_rf = RandomForestClassifier()
clf_rf.fit(X_train,y_train)
print('Elapsed time: ',str(dt.datetime.now()-start))
```

```python
clf_rf_cm=confusion_matrix(y_val.values, clf_rf.predict(X_val))

plt.title("Random Forest")
sns.heatmap(clf_rf_cm, annot = True, cmap="Blues",cbar=False);
```

```python
print("Random Forest \n")
print(mt.classification_report(y_val, clf_rf.predict(X_val),digits=3), "\n")
```

## Optimisation

## XGBoost

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
import time

clf_xgb = XGBClassifier()

max_depth= range(5,8)
learning_rate = [0.1, 0.2, 0.3]
colsample_bytree = [0.2,0.3,0.4]
reg_alpha = [0.7,0.8,0.9]
param_grid = dict(max_depth=max_depth,learning_rate=learning_rate,colsample_bytree=colsample_bytree,reg_alpha=reg_alpha)

grid = GridSearchCV(estimator=clf_xgb, param_grid=param_grid, cv = 3, n_jobs=-1)

start_time = time.time()
grid_result = grid.fit( X_train, y_train)
# Summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
print("Execution time: " + str((time.time() - start_time)) + ' ms')
```

` Train model again with tuned parameters and test on test dataset`

```python
from xgboost import XGBClassifier

start = dt.datetime.now()
clf_xgb = XGBClassifier(colsample_bytree= 0.4, learning_rate= 0.1, max_depth= 7, reg_alpha= 0.7)
clf_xgb.fit(X_train, y_train)
print('Elapsed time: ',str(dt.datetime.now()-start))
```

## Random Forest

```python
import time

criterion = ['entropy','gini']
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 50, stop = 400, num = 5)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 5)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap,
               'criterion':criterion}

clf_rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = clf_rf, param_distributions = random_grid, cv = 3, verbose=2, random_state=42, n_jobs = -1)

start_time = time.time()

# Fit the random search model
random_result = rf_random.fit(X_train, y_train)
print("Best: %f using %s" % (random_result.best_score_, random_result.best_params_))
print("Execution time: " + str((time.time() - start_time)) + ' ms')
```

```python
start = dt.datetime.now()
clf_rf = RandomForestClassifier(n_estimators= 50, min_samples_split= 5, min_samples_leaf= 2, max_features= 'sqrt', max_depth= None, bootstrap= False, criterion ="gini")
clf_rf.fit(X_train,y_train)
print('Elapsed time: ',str(dt.datetime.now()-start))
```

```python
print("Random Forest \n")
print(mt.classification_report(y_test, clf_rf.predict(X_test),digits=3), "\n")
```

```python
rest_of_xtest = X_test.iloc[:,7:]
rest_of_xtest
```

```python
#join back the rest of xtest
signal = signal.join(rest_of_xtest)
signal
```
