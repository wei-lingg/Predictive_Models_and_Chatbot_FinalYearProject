```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Opinions_Raw.csv") 
df.head()
```

```python
df.columns
```

```python
#Drop timestamp
df.drop(df.columns[0], axis=1, inplace=True)
df.columns
```

```python
#Rename column headers
df.columns = ['Age', 'Occupation','Annual Income','Covered with Policy?','Type of Policy covered','Policy bought by whom','Interest in buying another policy','Reason for being interested in buying another policy','Criteria looked out for buying insurance','Reason for being uninterested in buying another policy','Areas of coverage','Situation in which insurance will help you','Heard of Prudential?','Have policy under Prudential?','Type of policies under Prudential','Buy again?','Improvements']
df.columns
```

```python
#Convert all columns to lower case
df = df.apply(lambda x: x.astype(str).str.lower())
```

```python
#remove left and right spaces in occupation
df["Occupation"] = df["Occupation"].str.strip()
df["Occupation"].head()
```

```python
# See the different types of occupation
df.groupby('Occupation')['Occupation'].nunique()
```

```python
# Group student responses in "Student"
df.Occupation = df.Occupation.apply(lambda x: 'student' if 'stu' in x else x)
```

```python
df.head()
```

```python
df = df[(df.Age=='22 to 36 years old') | (df.Age =='21 years old and under') ]
```

## Extract values

```python
unique_list = ['life', 'motor/vehicle', 'health', 'savings', 'travel', 'investment', 'not sure']

s = "_PC"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    df.loc[df['Type of Policy covered'].str.contains(unique_list[i]), new_list[i]] = 1
    df[new_list[i]] = df[new_list[i]].fillna(0)

df.head()
```

```python
# Policy bought by whom
unique_list = ['myself', 'parents', 'government', 'school']


s = "_PB"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    # create new columns based on new insurance list
    df.loc[df['Policy bought by whom'].str.contains(unique_list[i]), new_list[i]] = 1
    #change Nan to 0
    df[new_list[i]] = df[new_list[i]].fillna(0)

df.head()
```

```python
# Reason for being interested
unique_list = ['future precautions', 'investment purposes']


s = "_RInt"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    # create new columns based on new insurance list
    df.loc[df['Reason for being interested in buying another policy'].str.contains(unique_list[i]), new_list[i]] = 1
    #change Nan to 0
    df[new_list[i]] = df[new_list[i]].fillna(0)

df['Reason for being interested in buying another policy'] = df['Reason for being interested in buying another policy'].fillna('NIL')
df.head()
```

```python
# Criteria looked out for
unique_list = ['affordability', 'coverage', 'clarity of policy', 'multi policy discounts', 'customer service'
              , 'convenience']


s = "_CL"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    # create new columns based on new insurance list
    df.loc[df['Criteria looked out for buying insurance'].str.contains(unique_list[i]), new_list[i]] = 1
    #change Nan to 0
    df[new_list[i]] = df[new_list[i]].fillna(0)

df['Criteria looked out for buying insurance'] = df['Criteria looked out for buying insurance'].fillna('NIL')
df.head()
```

```python
# Reason for being uninterested
unique_list = ['financial difficulties', 'lack of information',
                'unnecessary', 'family member']


s = "_RUint"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    # create new columns based on new insurance list
    df.loc[df['Reason for being uninterested in buying another policy'].str.contains(unique_list[i]), new_list[i]] = 1
    #change Nan to 0
    df[new_list[i]] = df[new_list[i]].fillna(0)

df['Reason for being uninterested in buying another policy'] = df['Reason for being uninterested in buying another policy'].fillna('NIL')
df.head()
```

```python
# Areas of coverage
unique_list = ['sickness', 'death', 'accident', 'travel', 'pandemic', 'disability', 'fraud', 'housing/property']


s = "_AC"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    # create new columns based on new insurance list
    df.loc[df['Areas of coverage'].str.contains(unique_list[i]), new_list[i]] = 1
    #change Nan to 0
    df[new_list[i]] = df[new_list[i]].fillna(0)

df.head()
```

```python
# Type of policies under Prudential
unique_list = ['life', 'motor/vehicle', 'health','savings', 'travel', 'investment', 
                  'nil', 'not sure', 'accident', 'critical illness', 'personal accident']

s = "_TPP"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    # create new columns based on new insurance list
    df.loc[df['Type of policies under Prudential'].str.contains(unique_list[i]), new_list[i]] = 1
    #change Nan to 0
    df[new_list[i]] = df[new_list[i]].fillna(0)

df.head()
```

## Further Cleaning for Text Analysis

```python
irrelevant_words = ['nil', 'nan', 'na', 'not sure', 'none', '-', 'idk', 'dk', 'ascacsa', 'dkkkkkk']


for i in range(len(irrelevant_words)):
    df['Improvements'] = df['Improvements'].replace(irrelevant_words[i],np.nan)
    df['Situation in which insurance will help you'] = df['Situation in which insurance will help you'].replace(irrelevant_words[i],np.nan)
    
df.head()
```

```python
pd.set_option('display.max_colwidth', -1)
display(df[['Improvements', 'Situation in which insurance will help you']].tail())
```

```python
# Remove punctuations, next line & digits

# Improvements
df.Improvements = df.Improvements.str.replace('[^\w\s]','')
# Situation
df['Situation in which insurance will help you'] = df['Situation in which insurance will help you'].str.replace('[^\w\s]','')
df['Situation in which insurance will help you'] = df['Situation in which insurance will help you'].str.replace('\n','')
df['Situation in which insurance will help you'] = df['Situation in which insurance will help you'].str.replace('\d+', '')

display(df[['Improvements', 'Situation in which insurance will help you']].tail())
```

```python
df = df.replace(np.nan, '', regex=True)
df = df.replace('nan', '', regex=True)
df.to_csv('Opinions_All_Cleaned.csv', index=False)
```
