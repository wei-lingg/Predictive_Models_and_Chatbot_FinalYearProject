```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ILP_Raw.csv") 
df.head()
```

```python
df.info()
```

```python
#Drop irrelevant columns
df.drop(df.iloc[:, 10:16], inplace = True, axis = 1) 
df.drop(df.iloc[:, 18:24], inplace = True, axis = 1) 
df.drop(df.iloc[:, 22:28], inplace = True, axis = 1) 
df.drop(df.iloc[:, 24:26], inplace = True, axis = 1) 
df.drop(df.iloc[:, 0:1], inplace = True, axis = 1) 
df.info()
```

```python
#Change column header name
df.columns = ['Age', 'Student?','Annual_Income','Purchased_before?','Reason_for_Purchasing','Channels_Used'
              ,'Rank_Channel_Used1','Rank_Channel_Used2','Rank_Channel_Used3'
              , 'Improvements', 'Reason_for_Not_Purchasing', 'Interest in ILP?', 'Reason_No_Interest', 'Lack_Info_Channel'
              ,'Rank_Channel_Lack_Wish1','Rank_Channel_Lack_Wish2','Rank_Channel_Lack_Wish3','Channels_Wish'
              ,'Rank_Channel_Wish1','Rank_Channel_Wish2','Rank_Channel_Wish3'
              , 'ILP_Opinions', 'Pru_ILP']
df.head()
```

```python
df = df.replace(np.nan, '', regex=True)
```

```python
#Convert all columns to lower case
df = df.apply(lambda x: x.astype(str).str.lower())

df.head()
```

## Extract values

```python
# Reason purchase
options_list = ['higher potential returns'
                , 'free switching of funds (get to choose the fund you invest in, allows you to optimise your investment returns by switching your investment to another fund)'
                , 'premium holidays (can temporarily stop premium payments if one is having financial difficulties)'
                , 'liquidity with partial withdrawals (without cancelling policy) & top ups (ad hoc basis)'
               , 'flexibility in insurance coverage'
               ]


unique_list = ['higher potential returns'
               , 'liquidity with partial withdrawals'
               , 'flexibility in insurance coverage'
               , 'free switching of funds'
               , 'premium holidays'
               ]

s = "_RP"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    df.loc[df['Reason_for_Purchasing'].str.contains(unique_list[i]), new_list[i]] = '1'
    df[new_list[i]] = df[new_list[i]].fillna(0)

df.head()


# We have to separate Others option for these columns 
# - Reasons_for_Purchasing
# - Reasons_for_Not_Purchasing
# - Reason_No_Interest
df['Others_RP'] = df['Reason_for_Purchasing']
df['Others_RP'] = df['Others_RP'].astype(str)
for i in range (len(options_list)):
    df['Others_RP'] = df['Others_RP'].apply(lambda x: x.replace(options_list[i],"",) if options_list[i] in x else x)
    df['Others_RP'] = df['Others_RP'].apply(lambda x: x.replace(";","") if ';' in x else x)
```

```python
# Reason not purchase
options_list = ['no guaranteed returns'
               ,'hefty fees'
               , 'possibility of reducing insurance coverage in the future'
               , 'complexity of the plans and their charges'
               , 'unnecessary'
               , 'not enough money'
               , 'lack of information']

unique_list = ['no guaranteed returns'
               ,'hefty fees'
               , 'possibility of reducing insurance coverage in the future'
               , 'complexity of the plans and their charges'
               , 'unnecessary'
               , 'not enough money'
               , 'lack of information']


s = "_RNP"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    # create new columns based on new insurance list
    df.loc[df['Reason_for_Not_Purchasing'].str.contains(unique_list[i]), new_list[i]] = 1
    #change Nan to 0
    df[new_list[i]] = df[new_list[i]].fillna(0)

# We have to separate Others option for these columns 
# - Reasons_for_Purchasing
# - Reasons_for_Not_Purchasing
# - Reason_No_Interest
df['Others_RNP'] = df['Reason_for_Not_Purchasing']
df['Others_RNP'] = df['Others_RNP'].astype(str)
for i in range (len(options_list)):
    df['Others_RNP'] = df['Others_RNP'].apply(lambda x: x.replace(options_list[i],"",) if options_list[i] in x else x)
    df['Others_RNP'] = df['Others_RNP'].apply(lambda x: x.replace(";","") if ';' in x else x)
```

```python
# Reason_No_Interest
options_list = ['lack of information'
                , 'unnecessary'
                , 'financial difficulties'
                , 'not willing to take the risk']

unique_list = ['lack of information'
                , 'unnecessary'
                , 'financial difficulties'
                , 'not willing to take the risk']


s = "_RNInt"
new_list = []
for i in range(len(unique_list)):
    new_list.append(unique_list[i]+s)

for i in range (len(unique_list)):
    # create new columns based on new insurance list
    df.loc[df['Reason_No_Interest'].str.contains(unique_list[i]), new_list[i]] = 1
    #change Nan to 0
    df[new_list[i]] = df[new_list[i]].fillna(0)
    
# We have to separate Others option for these columns 
# - Reasons_for_Purchasing
# - Reasons_for_Not_Purchasing
# - Reason_No_Interest
df['Others_RNInt'] = df['Reason_No_Interest']
df['Others_RNInt'] = df['Others_RNInt'].astype(str)
for i in range (len(options_list)):
    df['Others_RNInt'] = df['Others_RNInt'].apply(lambda x: x.replace(options_list[i],"",) if options_list[i] in x else x)
    df['Others_RNInt'] = df['Others_RNInt'].apply(lambda x: x.replace(";","") if ';' in x else x)
```
